```python
# Import necessary libraries
from astropy.table import Table
import pandas as pd
```

```python

# Define the file path
file_path = r"C:\git_repo\cool-dwarf_stellar_parameter_inference_from_survey_data\data\LAMOST_data\LAMOST_G0-9.vot"

# Read the .vot file using astropy
votable = Table.read(file_path, format='votable')

# Convert to pandas DataFrame
dataframe = votable.to_pandas()

# Display the first few rows of the DataFrame
dataframe.head()
```

```python
import numpy as np

# Divide the dataframe into 6 equally sized batches
batch_size = len(dataframe) // 6
remainder = len(dataframe) % 6

# Split the dataframe into 6 parts
batches = []
start_idx = 0
for i in range(6):
    # Distribute remainder rows among first batches
    end_idx = start_idx + batch_size + (1 if i < remainder else 0)
    batches.append(dataframe.iloc[start_idx:end_idx])
    start_idx = end_idx

# Convert back to astropy Tables and save as .vot files
for i, batch in enumerate(batches, 1):
    Table.from_pandas(batch).write(
        rf"C:\git_repo\cool-dwarf_stellar_parameter_inference_from_survey_data\data\LAMOST_data\LAMOST_G0-9_batch{i}.vot",
        format='votable',
        overwrite=True
    )
    print(f"Batch {i}: {len(batch)} rows")
```

```python
from astropy.table import Table
import pandas as pd
from concurrent.futures import ThreadPoolExecutor

# Define the file paths for the 6 batches
file_paths = [
    r"C:\git_repo\cool-dwarf_stellar_parameter_inference_from_survey_data\data\LAMOST_data\LAMOST_G0-9_Gaia_joined_batch1.vot",
    r"C:\git_repo\cool-dwarf_stellar_parameter_inference_from_survey_data\data\LAMOST_data\LAMOST_G0-9_Gaia_joined_batch2.vot",
    r"C:\git_repo\cool-dwarf_stellar_parameter_inference_from_survey_data\data\LAMOST_data\LAMOST_G0-9_Gaia_joined_batch3.vot",
    r"C:\git_repo\cool-dwarf_stellar_parameter_inference_from_survey_data\data\LAMOST_data\LAMOST_G0-9_Gaia_joined_batch4.vot",
    r"C:\git_repo\cool-dwarf_stellar_parameter_inference_from_survey_data\data\LAMOST_data\LAMOST_G0-9_Gaia_joined_batch5.vot",
    r"C:\git_repo\cool-dwarf_stellar_parameter_inference_from_survey_data\data\LAMOST_data\LAMOST_G0-9_Gaia_joined_batch6.vot"
]

# Function to read a single file
def read_vot_file(file_path):
    return Table.read(file_path, format='votable').to_pandas()

# Read files in parallel using ThreadPoolExecutor
with ThreadPoolExecutor(max_workers=8) as executor:
    dataframes = list(executor.map(read_vot_file, file_paths))
```

```python
combined_dataframe = pd.concat(dataframes, ignore_index=True)
print(f"Combined DataFrame shape: {combined_dataframe.shape}")
combined_dataframe.head()
```

```python
combined_dataframe.head(20)
```

```python
# Rename columns
combined_dataframe.rename(columns={
    'COLID_92974': 'obsid',
    'COLID_93001': 'gaia_dr3_id'
}, inplace=True)

# Remove duplicate columns (keep first occurrence)
combined_dataframe = combined_dataframe.loc[:, ~combined_dataframe.columns.duplicated()]

# Preserve only `obsid`, `gaia_dr3_id` and other requested columns
combined_dataframe = combined_dataframe[['obsid', 'gaia_dr3_id', 'Plx', 'Gmag', 'BPmag', 'RPmag']]
```

```python
combined_dataframe.head()
```

```python

# Convert the combined dataframe back to an astropy Table and save as a .vot file
output_file_path = r"C:\git_repo\cool-dwarf_stellar_parameter_inference_from_survey_data\data\LAMOST_data\LAMOST_G0-9_Gaia_joined.vot"
Table.from_pandas(combined_dataframe).write(output_file_path, format='votable', overwrite=True)

print(f"Combined dataframe saved to {output_file_path}")
```
