```python
import pandas as pd

path = r"C:\ASTR199\data\LAMOST_dM0-7.csv"
cols = ["mag_ps_g", "mag_ps_r", "mag_ps_i", "mag_ps_y", "mag_ps_z"]

df = pd.read_csv. I (path)

mask_999 = df[cols].isin([-999]).any(axis=1)
mask_na = df[cols].isna().any(axis=1)

count_999 = mask_999.sum()
count_na = (~mask_999 & mask_na).sum()  # missing only, excluding rows already flagged by -999

df = df[~(mask_999 | mask_na)]

print("dropped -999:", count_999)
print("dropped missing:", count_na)
```
