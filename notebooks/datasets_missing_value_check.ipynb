```python
import astropy.io.votable as votable
from astropy.table import Table
import numpy as np
```

```python
file_path = r"C:\git_repo\cool-dwarf_stellar_parameter_inference_from_survey_data\data\LAMOST_data\LAMOST_F0_2-9_Gaia_joined.vot"
vot = votable.parse(file_path)
table = vot.get_first_table().to_table()
print(f"Number of rows: {len(table)}")
print(f"Number of columns: {len(table.colnames)}")
print("Column names:", table.colnames)
```

```python
print("Checking for missing values in each column:")
for col in table.colnames:
    if hasattr(table[col], 'mask'):
        missing = np.sum(table[col].mask)
    else:
        # For non-masked columns, check for NaN in floats or None in objects
        if table[col].dtype.kind == 'f':  # float
            missing = np.sum(np.isnan(table[col]))
        elif table[col].dtype.kind == 'O':  # object (strings, etc.)
            missing = sum(1 for x in table[col] if x is None or str(x).strip() == '')
        else:
            missing = 0  # integers, etc., assume no missing
    print(f"{col}: {missing} missing values")
```
