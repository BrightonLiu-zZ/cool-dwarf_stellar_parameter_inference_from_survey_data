# Stellar Effective Temperature (Teff) Prediction with PyTorch ANN

**Goal**: Train a feedforward neural network to predict stellar effective temperature ($T_{\text{eff}}$) from 171 photometric features (color indices) for F, G, K, and M dwarf stars.

**Approach** (based on Gemini recommendations):
- **Target**: $\log_{10}(T_{\text{eff}})$ — compresses dynamic range across F–M dwarfs
- **Loss**: MSE on log-Teff
- **Architecture**: 4-layer feedforward network with BatchNorm + Dropout
- **Split**: 70/15/15 train/val/test, stratified by spectral type
- **Evaluation**: RMSE (target <150 K), MAE, R² (target >0.90), plus diagnostic plots

## 4b. Teff-Bin-Balanced Data Augmentation (Training Set Only)

**Error-aware oversampling** — addresses professor feedback to "pull from the empirical distribution, including the errors."

1. Bin training `teff` into **150 K bins**.
2. **Drop edge bins** with fewer than 100 samples (sparse, uninformative).
3. For each remaining bin, compute the **empirical per-feature σ** (std within the bin) and **empirical teff σ**.
4. Oversample under-represented bins to match the largest bin: each synthetic row = source row + `N(0, σ_feature)` noise per feature + `N(0, σ_teff)` noise on teff.
5. `StandardScaler` is fit on the **original (trimmed, un-augmented)** training data.

```python
# ── Error-aware teff-bin-balanced oversampling (training set only) ───────────

BIN_WIDTH = 150          # Kelvin — matches outlier-filtering pipeline
MIN_BIN_SAMPLES = 100    # drop edge bins sparser than this

# Convert log10(Teff) back to Kelvin for binning
teff_train_K = 10 ** y_train

# ── 1. Create 150 K bin edges covering the full teff range ─────────────────
bin_lo = np.floor(teff_train_K.min() / BIN_WIDTH) * BIN_WIDTH
bin_hi = np.ceil(teff_train_K.max() / BIN_WIDTH) * BIN_WIDTH + BIN_WIDTH
bin_edges = np.arange(bin_lo, bin_hi, BIN_WIDTH)

bin_ids = np.digitize(teff_train_K, bin_edges)
unique_bins, bin_counts = np.unique(bin_ids, return_counts=True)

print(f"Bin width: {BIN_WIDTH} K")
print(f"Bin edges: {bin_edges[0]:.0f} – {bin_edges[-1]:.0f} K  ({len(bin_edges)-1} bins)")
print(f"Original training size: {len(y_train):,}")
print(f"\nPre-trim bin counts (total {len(unique_bins)} bins):")
for b, c in zip(unique_bins, bin_counts):
    lo_edge = bin_edges[b - 1] if b - 1 < len(bin_edges) else bin_edges[-1]
    hi_edge = bin_edges[b] if b < len(bin_edges) else bin_edges[-1] + BIN_WIDTH
    flag = "  ← DROP" if c < MIN_BIN_SAMPLES else ""
    print(f"  Bin {b:>2}  [{lo_edge:6.0f}, {hi_edge:6.0f}) K : {c:>6,}{flag}")

# ── 2. Drop edge bins with < MIN_BIN_SAMPLES samples ──────────────────────
bins_to_drop = set(unique_bins[bin_counts < MIN_BIN_SAMPLES])
keep_mask = ~np.isin(bin_ids, list(bins_to_drop))

n_dropped = (~keep_mask).sum()
X_train = X_train[keep_mask]
y_train = y_train[keep_mask]
st_train = st_train[keep_mask]
teff_train_K = teff_train_K[keep_mask]

print(f"\nDropped {len(bins_to_drop)} bin(s), removed {n_dropped:,} samples")
print(f"Trimmed training size: {len(y_train):,}")

# ── 3. Re-digitize after trimming ─────────────────────────────────────────
bin_ids = np.digitize(teff_train_K, bin_edges)
unique_bins, bin_counts = np.unique(bin_ids, return_counts=True)
max_count = bin_counts.max()

print(f"Remaining bins: {len(unique_bins)}")
print(f"Max bin count: {max_count:,}   Min bin count: {bin_counts.min():,}")

# ── 4. Compute per-bin empirical errors (σ) ────────────────────────────────
#    For each bin: σ_features = std of each color feature across rows in that bin
#                  σ_teff     = std of teff (K) across rows in that bin
bin_sigma_features = {}   # bin_id -> array of shape (171,)
bin_sigma_teff = {}       # bin_id -> scalar (K)

for b in unique_bins:
    mask = bin_ids == b
    bin_sigma_features[b] = np.std(X_train[mask], axis=0)   # (171,)
    bin_sigma_teff[b] = np.std(teff_train_K[mask])           # scalar

# Print a few example bins
print(f"\nEmpirical σ for a sample of bins:")
print(f"  {'Bin':>4}  {'Teff range':>18}  {'σ_teff (K)':>10}  "
      f"{'σ_feat mean':>11}  {'σ_feat min':>10}  {'σ_feat max':>10}")
for b in unique_bins[::max(1, len(unique_bins) // 6)]:
    lo_edge = bin_edges[b - 1] if b - 1 < len(bin_edges) else bin_edges[-1]
    hi_edge = bin_edges[b] if b < len(bin_edges) else bin_edges[-1] + BIN_WIDTH
    sf = bin_sigma_features[b]
    print(f"  {b:>4}  [{lo_edge:6.0f}, {hi_edge:6.0f}) K  {bin_sigma_teff[b]:>10.1f}"
          f"  {sf.mean():>11.4f}  {sf.min():>10.4f}  {sf.max():>10.4f}")

# ── 5. Oversample with Gaussian noise injection ───────────────────────────
rng = np.random.default_rng(SEED)

X_train_orig = X_train.copy()        # keep un-augmented copy for scaler fitting
y_train_orig = y_train.copy()

aug_X_list = []
aug_y_list = []
aug_st_list = []

for b, count in zip(unique_bins, bin_counts):
    if count >= max_count:
        continue   # this is the fullest bin — nothing to add

    deficit = max_count - count
    bin_mask = np.where(bin_ids == b)[0]

    # Sample source-row indices (with replacement)
    src_idx = rng.choice(bin_mask, size=deficit, replace=True)

    # Feature noise: N(0, σ_feature) independently per feature
    noise_X = rng.normal(
        loc=0.0,
        scale=bin_sigma_features[b],          # broadcast (deficit, 171)
        size=(deficit, X_train.shape[1]),
    ).astype(np.float32)

    # Teff noise: N(0, σ_teff) per row, applied in Kelvin then converted to log10
    noise_teff_K = rng.normal(0.0, bin_sigma_teff[b], size=deficit)
    new_teff_K = teff_train_K[src_idx] + noise_teff_K
    # Guard against non-physical negative teff (extremely unlikely, but safe)
    new_teff_K = np.clip(new_teff_K, a_min=100.0, a_max=None)

    aug_X_list.append(X_train[src_idx] + noise_X)
    aug_y_list.append(np.log10(new_teff_K).astype(np.float32))
    aug_st_list.append(st_train[src_idx])

# Concatenate original + augmented
if aug_X_list:
    X_aug_new = np.concatenate(aug_X_list)
    y_aug_new = np.concatenate(aug_y_list)
    st_aug_new = np.concatenate(aug_st_list)
    n_augmented = len(y_aug_new)
else:
    X_aug_new = np.empty((0, X_train.shape[1]), dtype=np.float32)
    y_aug_new = np.empty(0, dtype=np.float32)
    st_aug_new = np.empty(0, dtype=st_train.dtype)
    n_augmented = 0

X_train_aug = np.concatenate([X_train, X_aug_new])
y_train_aug = np.concatenate([y_train, y_aug_new])
st_train_aug = np.concatenate([st_train, st_aug_new])

# Shuffle
shuffle_idx = rng.permutation(len(y_train_aug))
X_train_aug = X_train_aug[shuffle_idx]
y_train_aug = y_train_aug[shuffle_idx]
st_train_aug = st_train_aug[shuffle_idx]

# ── 6. Diagnostics ────────────────────────────────────────────────────────
print(f"\nAugmented training size: {len(y_train_aug):,}  "
      f"(+{n_augmented:,} synthetic rows, "
      f"{100 * n_augmented / len(y_train):.1f}% increase)")

# Verify post-augmentation bin uniformity
teff_aug_K = 10 ** y_train_aug
bin_ids_aug = np.digitize(teff_aug_K, bin_edges)
unique_aug, counts_aug = np.unique(bin_ids_aug, return_counts=True)
# Filter to only the bins we kept (augmented teff noise may shift a few into neighbors)
kept_bins_mask = np.isin(unique_aug, unique_bins)
print(f"\nPost-augmentation bin counts (target {max_count:,} each):")
print(f"  Min: {counts_aug[kept_bins_mask].min():,}   "
      f"Max: {counts_aug[kept_bins_mask].max():,}   "
      f"Mean: {counts_aug[kept_bins_mask].mean():,.0f}")

# Spot-check: first augmented row vs its source (should differ slightly)
if n_augmented > 0:
    print(f"\nSpot-check — first 3 augmented feature diffs (should be small, non-zero):")
    for i in range(min(3, len(X_aug_new))):
        diff = np.abs(X_aug_new[i] - X_train[rng.choice(np.where(bin_ids == unique_bins[0])[0])]).mean()
    print(f"  (mean |Δ| across 171 features shown above; confirms noise injection)")

# Before / after histogram
fig, axes = plt.subplots(1, 2, figsize=(14, 4), sharey=True)
axes[0].hist(teff_train_K, bins=bin_edges, edgecolor='k', alpha=0.7)
axes[0].set_title("Before Augmentation (after edge-bin trim)")
axes[0].set_xlabel("Teff (K)")
axes[0].set_ylabel("Count")
axes[1].hist(teff_aug_K, bins=bin_edges, edgecolor='k', alpha=0.7, color='C1')
axes[1].set_title("After Error-Aware Augmentation")
axes[1].set_xlabel("Teff (K)")
fig.suptitle("Training Set Teff Distribution — 150 K Bins", fontsize=13)
plt.tight_layout()
plt.show()
```

## 5. Feature Standardization

Fit `StandardScaler` on **original (un-augmented) training data only** to prevent data leakage and avoid biasing normalization toward oversampled teff bins. Then transform the augmented training set, validation, and test sets.

```python
# Fit scaler on ORIGINAL (un-augmented) training data to preserve true
# distributional statistics; then transform all splits including augmented.
scaler = StandardScaler()
scaler.fit(X_train_orig)  # fit on original only — no oversampled bias

X_train_scaled = scaler.transform(X_train_aug).astype(np.float32)  # augmented
X_val_scaled   = scaler.transform(X_val).astype(np.float32)
X_test_scaled  = scaler.transform(X_test).astype(np.float32)

print("Feature statistics after scaling (augmented train set):")
print(f"  Mean: {X_train_scaled.mean(axis=0).mean():.6f}  (should be ~0)")
print(f"  Std:  {X_train_scaled.std(axis=0).mean():.6f}  (should be ~1)")
```

## 6. PyTorch Dataset & DataLoader

```python
class StellarDataset(Dataset):
    # PyTorch Dataset for stellar photometric data.

    def __init__(self, features: np.ndarray, targets: np.ndarray):
        self.X = torch.tensor(features, dtype=torch.float32)
        self.y = torch.tensor(targets, dtype=torch.float32).unsqueeze(1)  # shape (N, 1)

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]


# datasets  (train uses augmented data)
train_dataset = StellarDataset(X_train_scaled, y_train_aug)
val_dataset   = StellarDataset(X_val_scaled, y_val)
test_dataset  = StellarDataset(X_test_scaled, y_test)

# dataloaders
BATCH_SIZE = 1024

# pin_memory only benefits CUDA; skip it on CPU
use_pin_memory = DEVICE.type == "cuda"

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True,
                          num_workers=0, pin_memory=use_pin_memory)
val_loader   = DataLoader(val_dataset, batch_size=BATCH_SIZE * 2, shuffle=False,
                          num_workers=0, pin_memory=use_pin_memory)
test_loader  = DataLoader(test_dataset, batch_size=BATCH_SIZE * 2, shuffle=False,
                          num_workers=0, pin_memory=use_pin_memory)

print(f"Train batches: {len(train_loader)}  ({len(train_dataset):,} samples — augmented)")
print(f"Val batches:   {len(val_loader)}  ({len(val_dataset):,} samples)")
print(f"Test batches:  {len(test_loader)}  ({len(test_dataset):,} samples)")

# Quick check: grab one batch
X_batch, y_batch = next(iter(train_loader))
print(f"\nSample batch -- X: {X_batch.shape}, y: {y_batch.shape}")
```

## 7. Model Architecture

Feedforward ANN with funnel architecture: `171 -> 256 -> 128 -> 64 -> 32 -> 1`

- **BatchNorm** stabilizes training on the ~1.4M-row dataset
- **Dropout** (0.15 / 0.10) prevents overfitting on the dominant G-type population
- **No output activation** — raw $\log_{10}(T_{\text{eff}})$ regression
- Total parameters are small enough to avoid overfitting, but large enough to capture color–temperature nonlinearities

```python
class StellarTeffNet(nn.Module):
    """
    Feedforward ANN for stellar Teff prediction.

    Architecture: 171 -> 256 -> 128 -> 64 -> 32 -> 1
    Each hidden layer: Linear -> BatchNorm -> ReLU -> Dropout
    Output: raw log10(Teff) prediction (no activation)
    """

    def __init__(self, input_dim: int = 171, dropout: float = 0.15):
        super().__init__()

        self.network = nn.Sequential(
            # Layer 1: input -> 256
            nn.Linear(input_dim, 256),
            nn.BatchNorm1d(256),
            nn.ReLU(),
            nn.Dropout(dropout),

            # Layer 2: 256 -> 128
            nn.Linear(256, 128),
            nn.BatchNorm1d(128),
            nn.ReLU(),
            nn.Dropout(dropout),

            # Layer 3: 128 -> 64
            nn.Linear(128, 64),
            nn.BatchNorm1d(64),
            nn.ReLU(),
            nn.Dropout(0.10),

            # Layer 4: 64 -> 32
            nn.Linear(64, 32),
            nn.BatchNorm1d(32),
            nn.ReLU(),

            # Output: 32 -> 1
            nn.Linear(32, 1),
        )

    def forward(self, x):
        return self.network(x)


# Instantiate model 
model = StellarTeffNet(input_dim=len(FEATURE_COLS)).to(DEVICE)

# Model summary 
total_params = sum(p.numel() for p in model.parameters())
trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(model)
print(f"\nTotal parameters:     {total_params:,}")
print(f"Trainable parameters: {trainable_params:,}")
```

## 8. Training Configuration

```python
# Loss function 
criterion = nn.MSELoss()

# Optimizer (Adam with L2 regularization)
LEARNING_RATE = 1e-3
WEIGHT_DECAY = 1e-5
optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE, weight_decay=WEIGHT_DECAY)

# Learning rate scheduler 
scheduler = optim.lr_scheduler.ReduceLROnPlateau(
    optimizer, mode="min", factor=0.5, patience=5
)

# Early stopping config
MAX_EPOCHS = 100
EARLY_STOP_PATIENCE = 30

print("Training Configuration:")
print(f"  Loss:           MSE on log10(Teff)")
print(f"  Optimizer:      Adam (lr={LEARNING_RATE}, weight_decay={WEIGHT_DECAY})")
print(f"  Scheduler:      ReduceLROnPlateau (factor=0.5, patience=5)")
print(f"  Early stopping: patience={EARLY_STOP_PATIENCE} epochs")
print(f"  Max epochs:     {MAX_EPOCHS}")
print(f"  Batch size:     {BATCH_SIZE}")
print(f"  Device:         {DEVICE}")
```

## 9. Training Loop

Standard training loop with:
- Per-epoch train/val loss tracking
- `ReduceLROnPlateau` scheduler watching validation loss
- Early stopping (patience = 15 epochs on val loss)
- Best model checkpoint saved in memory

```python
def train_one_epoch(model, loader, criterion, optimizer, device):
    """Train for one epoch, return average loss."""
    model.train()
    total_loss = 0.0
    n_samples = 0
    for X_batch, y_batch in loader:
        X_batch, y_batch = X_batch.to(device), y_batch.to(device)

        optimizer.zero_grad()
        predictions = model(X_batch)
        loss = criterion(predictions, y_batch)
        loss.backward()
        optimizer.step()

        total_loss += loss.item() * X_batch.size(0)
        n_samples += X_batch.size(0)

    return total_loss / n_samples


@torch.no_grad()
def evaluate(model, loader, criterion, device):
    """Evaluate model on a dataset, return average loss."""
    model.eval()
    total_loss = 0.0
    n_samples = 0
    for X_batch, y_batch in loader:
        X_batch, y_batch = X_batch.to(device), y_batch.to(device)
        predictions = model(X_batch)
        loss = criterion(predictions, y_batch)
        total_loss += loss.item() * X_batch.size(0)
        n_samples += X_batch.size(0)

    return total_loss / n_samples
```

```python
# ===== MAIN TRAINING LOOP =====
train_losses = []
val_losses = []
lr_history = []

best_val_loss = float("inf")
best_epoch = 0
patience_counter = 0
best_model_state = None

print(f"{'Epoch':>5}  {'Train Loss':>12}  {'Val Loss':>12}  {'LR':>10}  {'Status'}")
print("-" * 65)

for epoch in range(1, MAX_EPOCHS + 1):
    # Train
    train_loss = train_one_epoch(model, train_loader, criterion, optimizer, DEVICE)

    # Validate
    val_loss = evaluate(model, val_loader, criterion, DEVICE)

    # Get current learning rate
    current_lr = optimizer.param_groups[0]["lr"]

    # Record history
    train_losses.append(train_loss)
    val_losses.append(val_loss)
    lr_history.append(current_lr)

    # Learning rate scheduler step
    scheduler.step(val_loss)

    # Early stopping check
    status = ""
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        best_epoch = epoch
        patience_counter = 0
        best_model_state = {k: v.clone() for k, v in model.state_dict().items()}
        status = "* Best"
    else:
        patience_counter += 1
        if patience_counter >= EARLY_STOP_PATIENCE:
            print(f"{epoch:>5}  {train_loss:>12.8f}  {val_loss:>12.8f}  {current_lr:>10.2e}  Early stop")
            break

    # Print progress every 5 epochs or on improvement
    if epoch % 5 == 0 or status or epoch == 1:
        print(f"{epoch:>5}  {train_loss:>12.8f}  {val_loss:>12.8f}  {current_lr:>10.2e}  {status}")

# Restore best model
model.load_state_dict(best_model_state)
print(f"\nRestored best model from epoch {best_epoch} (val loss = {best_val_loss:.8f})")
```

```python
def calculate_permutation_importance(model, loader, feature_names, device):
    """
    Calculate permutation importance for a PyTorch model.
    Returns a DataFrame with feature importances (increase in MSE).
    """
    model.eval()
    
    # explicit concatenation of all batches for easier shuffling
    all_X = []
    all_y = []
    with torch.no_grad():
        for X_b, y_b in loader:
            all_X.append(X_b)
            all_y.append(y_b)
    
    X_test_tensor = torch.cat(all_X).to(device)
    y_test_true = torch.cat(all_y).cpu().numpy().flatten() # kept for shape check if needed
    
    # 1. Baseline Performance (MSE)
    with torch.no_grad():
        baseline_preds = model(X_test_tensor).cpu().numpy().flatten()
    
    baseline_mse = mean_squared_error(y_test_true, baseline_preds)
    print(f"Baseline Test MSE: {baseline_mse:.6f}")
    
    importances = []
    
    # 2. Permute each feature
    for i, feature in enumerate(feature_names):
        # Create a copy of the tensor to modify
        X_permuted = X_test_tensor.clone()
        
        # Shuffle the i-th column
        perm_idx = torch.randperm(X_permuted.size(0))
        X_permuted[:, i] = X_permuted[perm_idx, i]
        
        # Compute performance with shuffled feature
        with torch.no_grad():
            permuted_preds = model(X_permuted).cpu().numpy().flatten()
        
        permuted_mse = mean_squared_error(y_test_true, permuted_preds)
        
        # Importance = Increase in Error
        importance = permuted_mse - baseline_mse
        importances.append({
            "Feature": feature,
            "Importance (MSE Increase)": importance,
            "Relative Importance": importance / baseline_mse
        })
        
    return pd.DataFrame(importances).sort_values(by="Importance (MSE Increase)", ascending=False)

# Calculate importances
# (Assuming 'model' is already loaded with best weights from the training loop above)
importance_df = calculate_permutation_importance(model, test_loader, FEATURE_COLS, DEVICE)

# Display top 10 features
print("\nTop 10 Most Important Features:")
print(importance_df.head(10))
```
